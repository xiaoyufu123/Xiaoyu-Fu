```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('../Downloads/amazon.csv', encoding='latin1')
print('Number of features: %s' %data.shape[1])
print('Number of examples: %s' %data.shape[0])
```

```python
data.head()
```

```python
data1 = pd.DataFrame(data.groupby('year')['number'].sum().reset_index())

plt.figure(figsize=(15,8))
sns.lineplot(x='year', y='number', data=data1, color='red', lw=2)
plt.xlabel('Year')
plt.ylabel('Number of fires')
plt.title('Number of fires in Brazil per year', fontsize=15)
plt.xlim(1998,2017)
plt.xticks(np.arange(1998, 2018, 1),fontsize=12)
plt.yticks(fontsize=12)
```

```python
data2 = pd.DataFrame(data.groupby('state')['number'].sum().sort_values(ascending=False).reset_index())

plt.figure(figsize=(15,8))
sns.barplot(data=data2, x='state', y='number', palette='summer')
plt.title('Number of fires per state', fontsize=15)
plt.xlabel('States')
plt.ylabel('Number of fires')
plt.xticks(rotation=45)
```
